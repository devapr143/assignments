# 11.In a for loop, what is the difference between range(10), range(0, 10), and range(0, 10, 1)?
Ans:The range (10) means it calls range from 0 to 9, The range (0,10) also means that it calls range from 0 to 9 with the starting value being 0, The range (0,10,1) means it calls range from 0 to 9 at the intervals of 1. They all do the same thing in different methods.
